```python
from matplotlib import pyplot as plt
import seaborn as sns
import pandas as pd

%matplotlib inline
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/fivethirtyeight/data/master/flying-etiquette-survey/flying-etiquette.csv')
```

```python
df.head()
```

```python
# Percent breakdown of survey participants by age range.

gb = df.Age.value_counts() / len(df) * 100
```

```python
gb
```

```python
x = gb.index
```

```python
y = gb.values
```

```python
age_order = ["18-29", "30-44", "45-60", "> 60"]

sns.countplot(x="Age", data = df, order = age_order)

plt.xlabel("Age Range")
plt.ylabel("Count")
plt.title("Breakdown of Survey Participants by Age Range")
```

```python
# 2. How did survey respondents generally feel regarding if “it [was] 
# rude to ask someone to switch seats with you in order to be closer to 
# friends?”?

df.columns
```

```python
rude = 'Is it rude to ask someone to switch seats with you in order to be closer to friends?'
```

```python
sns.countplot(x = rude, data = df)

plt.xlabel("Participant Response")
plt.ylabel("Count")
plt.title("Breakdown of Survey Participants by Opinion on Being ask to Switch Seats")
```

```python
# 3. Use the respondents age as a dimension to see if there are any 
# differences to responses to “it [was] rude to ask someone to switch 
#seats with you in order to be closer to friends?”?


sns.displot(data=df, y=rude, col='Age', col_wrap=2)

```
